# Batsmen vs Bowlers

```python
newSeries = deliveries[['batsman','bowler','batsman_runs','player_dismissed']]

AllBatsmen = newSeries.batsman.unique();

AllBowlers = newSeries.bowler.unique();

runsScored=[];
ballsFaced=[];
batsmenBowlerBalls = {'Batsman ' : '# Balls'}
batsmenBowler = {'Batsmen' : 'RunsScored'}

for batsmen in AllBatsmen :
    print(batsmen)
    for bowler in AllBowlers:
        tmpSeries=newSeries[(newSeries.batsman==batsmen) & (newSeries.bowler==bowler)].groupby('batsman').sum();
        ballsFaced.append(len(newSeries[(newSeries.batsman==batsmen) & (newSeries.bowler==bowler)]))
        if len(tmpSeries)==0:
            runsScored.append(0)
        else :
            tmpVal=newSeries[(newSeries.batsman==batsmen) & (newSeries.bowler==bowler)].groupby('batsman').sum().batsman_runs[0];
            runsScored.append(tmpVal);
            
    batsmenBowler[batsmen]=runsScored;
    batsmenBowlerBalls[batsmen]=ballsFaced;
    runsScored=[];
    ballsFaced=[];
```

```python
maxRuns = {batsmen : max(batsmenBowler[batsmen]) for batsmen in AllBatsmen}
```

```python
maxRuns
```

```python
maxBalls = {batsmen : max(batsmenBowlerBalls[batsmen]) for batsmen in AllBatsmen}
```

```python
maxRunsSorted=sorted(maxRuns,key=maxRuns.get,reverse=True);
maxBallsSorted=sorted(maxBalls,key=maxBalls.get,reverse=True);
```

```python
Player = 'Yuvraj Singh'
print('Runs Scored: {}'.format(maxRuns[Player]))

print('Bowler : {}'.format(AllBowlers[np.argmax(batsmenBowler[Player])]))

print('Balls faced : {}'.format(batsmenBowlerBalls[Player][np.argmax(batsmenBowler[Player])]))
```
